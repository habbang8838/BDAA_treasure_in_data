```python
# 데이터프레임은 시트가 여러개이므로, csv 파일로 불러오는 것이 불가능하여, 리스트에 요소별로 데이터프레임을 추가하고, concat하는 방식으로 하였습니다.
df = []
for i in range(9):
    df.append(pd.read_excel('BDAA_1회 내부 공모전 데이터.xlsx', sheet_name=i))
```

df_canceled 데이터프레임에 따로 저장할게요~!

```python
df['유형'].unique()
```

```python
df_canceled = df.loc[(df['구매확정'] == '구매미확정') | (df['유형'] == '반품') | (df['예외처리구분'].notna())]
```

```python
# 유형에 "주문"은 뭘까요?? 취소된 데이터일까요~~?

df_canceled['유형'].unique()
```

유형이 주문, 예외처리구분에 값이 없는 주문 3건은, 주문접수가 9월이며,  취소 여부가 불확실한 데이터군요! -> 취소한 제품들로 성급하게 판단하면 안될 거 같아요. 얘네는 df_canceled에서 빼야겠어요!

```python
df_canceled.loc[(df_canceled['유형'] == '주문') & (df_canceled['예외처리구분'].isna())]
```

- 아까 모았던 구매 미확정 데이터 중 유형이 '취소(주문취소)', '반품' 이거나, 유형은 '주문'인데 예외처리구분이 ('미수령신고접수철회요청', '구매확정후취소')인 데이터를 df_canceled에 모아주겠습니다.

```python
df_canceled = df_canceled.loc[(df_canceled['유형'] == '취소(주문취소)') | ((df_canceled['유형'] == '반품')) | ((df_canceled['예외처리구분'].notna()))]
```

- 전체 데이터 102077개중 주문 취소 및 구매 확정후 취소, 미수령신고접수 후 철회요청 된 데이터 8702개가 df_canceled에 저장되었군요!


```python
df_canceled
```

```python
df_canceled['판매액'] = df_canceled['판매가'] * df_canceled['수량']
```

```python
df_canceled['주문접수일시'] = pd.to_datetime(df_canceled['주문접수일시'], format = "%Y-%m-%d %H:%M:%S")
```

```python
df_canceled['주문월'] = df_canceled['주문접수일시'].dt.month
```

```python
df_canceled = df_canceled.reset_index()
```

```python
df_canceled = df_canceled.drop(['index'], axis=1)
```
