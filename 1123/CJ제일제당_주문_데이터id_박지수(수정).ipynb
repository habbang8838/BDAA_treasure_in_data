```python
# 먼저 id가 3회 이상 나온 데이터에 한해서만 저장하도록 할게요!!

df_up3 = df[df['아이디'] == df['아이디'].value_counts().loc[lambda x: x>= 3].index[0]]
for i in range(1, len(df['아이디'].value_counts().loc[lambda x: x>= 3].index)):
    df_up3 = df_up3.append(df[df['아이디'] == df['아이디'].value_counts().loc[lambda x: x>= 3].index[i]])
df_up3
```

```python
# 일단 아이디가 저렇게 *로 되거나 null 값인 경우는 같은 회원으로 보기 어려워 삭제해줄게요~~

df_up3 = df_up3[(df_up3['아이디'] != '*') & (df_up3['아이디'] != '**') & 
                  (df_up3['아이디'] != '***') & (df_up3['아이디'] != '****') & (df_up3['아이디'] != '')]
```

```python
df_up3['시간대'] = df_up3['주문접수일시'].dt.hour
```
