```python
df_up5 = df[df['아이디'] == df['아이디'].value_counts().loc[lambda x: x>= 5].index[0]]
for i in range(1, len(df['아이디'].value_counts().loc[lambda x: x>= 5].index)):
    df_up5 = df_up5.append(df[df['아이디'] == df['아이디'].value_counts().loc[lambda x: x>= 5].index[i]])
df_up5
```

```python
df_up5 = df_up5[(df_up5['아이디'] != '*') & (df_up5['아이디'] != '**') & 
                  (df_up5['아이디'] != '***') & (df_up5['아이디'] != '****') & (df_up5['아이디'] != '')]
```

```python
df_up5_test = df_up5.groupby(['아이디', '판매자상품명']).count()
```

```python
df_up5_cond = df_up5_test[df_up5_test['주문접수일시'] >= 2]
```
